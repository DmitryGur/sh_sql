```python
# код для получения списка городов-милионников

import requests
from bs4 import BeautifulSoup
import re

subject = 'Города-миллионеры_России'
url = 'https://ru.wikipedia.org/w/api.php'

params = {
    'action': 'parse',           # Используем действие "parse"
    'format': 'json',
    'page': subject,
    'prop': 'text',             # Получаем HTML-текст страницы
}
#Любой header, без него запрос не принимается
headers = {
    'User-Agent': 'MyPythonScript/1.0 (your_email@example.com)'
}

response = requests.get(url, params=params, headers=headers)
response.encoding = 'utf-8'

data = response.json()
#К сожалению, таблица формируется JS и не передается в json, поэтому используем страницу в виде html
# Извлекаем HTML-код из ответа
html_content = data['parse']['text']['*']

# Создаем объект BeautifulSoup для парсинга HTML
soup = BeautifulSoup(html_content, 'html.parser')
#Найдем таблицу
target_table = soup.find('tbody')
#С помощью regexp получим все заголовки ссылок
cities = re.findall(r'title=("[\w -]+")', str(target_table))
#Уберем лишние ковычки
cities = [city.strip('"') for city in cities]
print(cities)
```

```python
# текст запроса

query_5_7 = '''
WITH yandex_vacancies AS (
    SELECT 
        areas.id
,areas.name
 AS area_name
    FROM
        vacancies 
    JOIN 
        employers ON vacancies.employer_id=employers.id
    JOIN 
        areas ON vacancies.area_id=areas.id
    WHERE 
        employers.name='Яндекс' 
        AND areas.name
 IN %s
    )
SELECT 
    area_name, COUNT(id) AS cnt 
FROM 
    yandex_vacancies
GROUP BY 
    area_name
UNION ALL
SELECT 
    'total' AS area_name, 
    COUNT(id) AS cnt
FROM 
    yandex_vacancies;
'''
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
# текст запроса
query_6_6= f'''WITH salary_calc AS (
    SELECT
        experience,
        (COALESCE(salary_from, salary_to) + COALESCE(salary_to, salary_from)) / 2 AS calculated_salary
    FROM vacancies
    WHERE 
        -- Фильтрация по DS-вакансиям
        (LOWER(name) LIKE '%data scientist%'
         OR LOWER(name) LIKE '%data science%'
         OR LOWER(name) LIKE '%исследователь данных%'
         OR (name LIKE '%ML%' AND LOWER(name) NOT LIKE '%html%')
         OR LOWER(name) LIKE '%machine learning%'
         OR LOWER(name) LIKE '%машинн%обучен%')
        
        -- Условие на наличие хотя бы одного заполненного поля зарплаты
        AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
)
SELECT
    experience,
    ROUND(AVG(calculated_salary)) AS avg_salary
FROM salary_calc
GROUP BY experience
'''
```

```python
# результат запроса
df_6_6 = pd.read_sql_query(query_6_6, connection)
df_6_6
```

Иссдедоване средней ЗП специалистов по данным в зависимости от названия должности.

```python
query_7_1= f'''WITH salary_calc AS (
    SELECT
        name,
        CASE 
            WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL THEN (salary_from + salary_to) / 2
            WHEN salary_from IS NOT NULL THEN salary_from
            WHEN salary_to IS NOT NULL THEN salary_to
        END AS calculated_salary
    FROM vacancies
    WHERE 
        -- Фильтрация по DS-вакансиям
        (LOWER(name) LIKE '%data scientist%'
         OR LOWER(name) LIKE '%data science%'
         OR LOWER(name) LIKE '%исследователь данных%'
         OR (LOWER(name) LIKE '%ML%' AND LOWER(name) NOT LIKE '%html%')
         OR LOWER(name) LIKE '%machine learning%'
         OR LOWER(name) LIKE '%машинн%обучен%')
        
        -- Условие на наличие хотя бы одного заполненного поля зарплаты
        AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
)
SELECT
    name,
    ROUND(AVG(calculated_salary)) AS avg_salary
FROM salary_calc
GROUP BY name
ORDER BY avg_salary DESC
'''
```
